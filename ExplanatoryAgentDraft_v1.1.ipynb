```python
# EXAMPLE QUERY TO GENETICS PROVIDER to obtain a list of genes

import json
gwas_url = 'https://translator.broadinstitute.org/genetics_data_provider/query'
query = """
{
  "message": {
    "query_graph": {
      "edges": [
        {
          "id": "e00",
          "source_id": "n00",
          "target_id": "n01",
          "type": "associated"
        }
      ],
      "nodes": [
        {
          "curie": "EFO:0004466",
          "id": "n00",
          "type": "disease"
        },
        {
          "id": "n01",
          "type": "gene"
        }
      ]
    }
  }
}
"""
query = json.loads(query)
```

```python
# EXECUTE QUERY TO OBTAIN RESPONSE
import requests
gwas_response = requests.post(gwas_url, json=query).json()
gwas_response
```

```python
# populates the output into a list called gene_ids
gene_ids = [node['id'] for node in gwas_response['knowledge_graph']['nodes'] if node['type']=='gene']
# this counts the number of items in the output
len(gene_ids)
```

```python
# GIVEN THAT WE HAVE THE LIST OF GENES, WE NEED TO LOOP INTO THE QUERY AND REPLACE 
# THE GENE CURIE FOR NODE 1 WITH EACH GENE FROM THE GENE LIST
# the list of genes is gene_ids[]
```

```python
import requests

url = 'https://translator.broadinstitute.org/molepro_reasoner/query'

#this is the fuction for the task
def ARA_Result_function(gene_id, type_id):
    query ={
        "message": {
            "query_graph": {
                "nodes": [
                    {
                        "curie": gene_id,
                        "id": "n00",
                        "type": "gene"
                    },
                    {
                        "id": "n01",
                        "type": "chemical_substance"
                    }
                ],
                "edges": [
                    {
                        "id": "e00",
                        "source_id": "n00",
                        "target_id": "n01",
                        "type": type_id
                    }
                ]
            }
        }
    }
    res = requests.post(url, json=query).json()
    for result in res['results']:
        result['score'] = "None"
        result['Explanatory_Agent_Ranking'] = 1 if type_id == 'affected_by' else 2 if type_id == 'correlated_with' else 3
    return res

# Workaround for not being able to batch query. We are treating the first 25 genes as if it were a batch query.
# Making adjustments to try and fit the example query response supplied from Mark Williams.
first_iteration = True
for type_id in [ 'affected_by', 'correlated_with', '' ]:  # Supported edge types for the use-case query
    for gene_id in gene_ids[0:150]:     #here we can limit the number of genes used to limit use of KP while testing
        if first_iteration:
            response = ARA_Result_function(gene_id, type_id)  # intialize core response graph
            first_iteration = False
        else:  # Extend the intial response graph with new responses found for each query iteration (Only if they contain responses).
            new_response = ARA_Result_function(gene_id, type_id)
            if new_response['results']:
                response['results'].extend(new_response['results'])
                response['knowledge_graph']['edges'].extend(new_response['knowledge_graph']['edges'])
                response['knowledge_graph']['nodes'].extend(new_response['knowledge_graph']['nodes'])

# workaround until we get a standard query graph message format from ARS     
response["query_graph"] = _query_graph["query_graph"]  
            

with open('looping_query_test.json', 'w') as outfile:
    json.dump(response, outfile)
```
